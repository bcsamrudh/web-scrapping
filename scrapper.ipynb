```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
reviews = soup.find_all(class_="_2-N8zT")
names = soup.find_all(class_="_2sc7ZR _2V5EHH")
```

```python
cust_name=set()
for name in names:
    cust_name.add(name.get_text().strip('\n'))
cust_name
```

```python
review_title=set()
review_title_cleaned =set()
for review in reviews:
    review_title.add(review.get_text())
# for i in review_title:
#     if '\n' in i:
#        review_title_cleaned.add(i.split('\n')[1])
#     else:
#         review_title_cleaned.add(i)
# review_title_cleaned
list(review_title)
```

```python
data = {
    "Name" : list(cust_name)[0:len(review_title)],
    "Review" : list(review_title)
}
df = pd.DataFrame(data)
writer = pd.ExcelWriter('Iphone_reviews.xlsx',engine='xlsxwriter')
df.to_excel(writer,sheet_name="List")
writer.save()
```
